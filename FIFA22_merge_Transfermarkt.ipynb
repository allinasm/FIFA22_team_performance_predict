```python
def get_tables_links_transfermarkt(n_pages):
    
    headers = {'User-Agent': 
           'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}
    
    
    url_links = {}
    
    for i in range(1,n_pages):
        r = requests.get('https://www.transfermarkt.es/wettbewerbe/europa?ajax=yw1&page='+str(i), headers=headers)
        # Obtener el contenido HTML de la página web.
        html = r.content
        #Crear el objeto soup
        soup = BeautifulSoup(html, 'lxml')
    
        #Obtener la tabl
        table = soup.find('table', attrs={'class': 'items'})
        #Obtener las filas
        trs = table.find("tbody").find_all('tr')
        # obtener solo los links
        for tr in trs:
            for i, cell in enumerate(tr.find_all('a')):
                if url_links.__contains__(cell.get('title')) == False:
                    url_links[cell.get('title')] = 'https://www.transfermarkt.es' + cell.get('href')
                    
    # Elimina algun valor vacio
    url_links.pop(None)
    
    table_links = {}

    for name, url in url_links.items():
        r = requests.get(url, headers=headers)
    
        # Obtener el contenido HTML de la página web.
        html = r.content
        #Crear el objeto soup
        soup = BeautifulSoup(html, 'lxml')
    
        #Obtener la tabl
        table = soup.find('a', attrs={'title': 'Clasificación completa'})
        if table != None:
            link = table.get('href')
            table_links[name] = 'https://www.transfermarkt.es' + link
    
    return table_links
```

```python
def get_table(url):
    
    headers = {'User-Agent': 
           'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}
    
    r = requests.get(url, headers=headers)
    
    # Obtener el contenido HTML de la página web.
    html = r.content
    # Crear el objeto soup
    soup = BeautifulSoup(html, 'lxml')
    
    #Obtener la tabla
    table = soup.find('table', attrs={'class': 'items'})
    
    # Obtener la cabecera de las columnas
    ths=table.find("thead").find_all('th')
    
    # Modificar algunos valores
    col_headings= [th.text.strip('\n').strip('\xa0') for th in ths]
    col_headings[2] ='PJ'
    col_headings.insert(1,'Shield')
    
    #Obtener las filas
    trs=table.find("tbody").find_all('tr')
    #Obtener las celdas por fila
    tds= [tr.find_all('td') for tr in trs]
    
    table=[]
    
    for i in range(0,len(tds)):
        table.append([tt.text.strip('\n').strip(' \xa0') for tt in tds[i]])
        
    df_tabla = pd.DataFrame(table,columns=col_headings)
    df_tabla[['GF', 'GC']] = df_tabla['Goles'].str.split(':',n=2, expand=True)
    df_tabla = df_tabla[["#", "Shield", "Club", "PJ", "G", "E", "P", "GF", "GC", "+/-", "Pto."]]
    df_tabla = df_tabla.astype({"#": int, "PJ": int, "G": int, "E": int, "P": int, "GF": int, "GC": int ,
                                "+/-": int, "Pto.":int,})
    
    return df_tabla
```
